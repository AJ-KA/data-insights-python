# 1. If-statements with user-defined variables

```python
# Create a subset

df = ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
# Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function and create column price_range

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

# 3. If-statements with For-Loops

```python
# Printing the frequency of the “orders_day_of_week” column, to know on which day most orders take place

ords_prods_merge['orders_day_of_the_week'].value_counts(dropna=False)
```

```python
print(ords_prods_merge.columns)
```

```python
# Create an empty list

result = []

for value in ords_prods_merge["orders_day_of_the_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create a new column in df and set it equal to result

ords_prods_merge['busiest_day'] = result
```

```python
# Print frequency of new column

ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

## 4. Task

#### 4.2 Creating new column busiest_days

```python
# Create an empty list

result2 = []

for value in ords_prods_merge["orders_day_of_the_week"]:
    if value in [0, 1]:
        result2.append("Busiest days")
    elif value in [3, 4]:
        result2.append("Least busy")
    else:
        result2.append("Regularly busy")
```

```python
# Create a new column in df and set it equal to result2

ords_prods_merge['busiest_days'] = result2
```

## 4.3. Check value of new column for accuracy

```python
result2
```

```python
# Print frequency of new column

ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# The frequency added up is the same count as the frequency of the busiest day column, which makes sense since we have only changed the classification of values but not the values themselves. 
```

## 4.4 Create new column busiest_period_of_day

```python
# Printing the frequency of the “order_hour_of_day” column, to know when in the day most orders take place

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating an empty list

result3 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result3.append("Most orders")
    elif value in [3, 4, 2, 5, 1, 0, 6, 23 ]:
        result3.append("Average orders")
    else:
        result3.append("Fewest orders")
```

```python
# Checking the result

result3
```

```python
# Create a new column in df and set it equal to result3

ords_prods_merge['busiest_period_of_day'] = result3
```

## 4.5 Print frequency of column

```python
# Print frequency of new column

ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

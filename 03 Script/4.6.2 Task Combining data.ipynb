```python
# import orders_products_combined data 

df_ords_prods_comb = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_combined.pkl'))
```

```python
# import products_checked data as df_prods

df_prods = pd.read_csv (os.path.join(path, '02 Data','Prepared Data', 'products_checked.csv'), index_col = 0)
```

### 3. Merge with updated products data set

```python
# test merge without overwriting with merge flag 'Test'. As we will be workign with merge rates that have a full merge rate the default inner join is chosen.

pd.merge(df_ords_prods_comb, df_prods, on = ['product_id'] , indicator = 'Test')
```

```python
# merging df_ords_prods_comb and df_prods using product_id as key 

ords_prods_merge = df_ords_prods_comb.merge(df_prods, on = 'product_id', indicator = 'Match')
```

### 4. Confirm merge results

```python
ords_prods_merge['Match'].value_counts()
```
